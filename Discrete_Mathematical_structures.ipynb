```python
from sympy import symbols, And, Or, Not, Implies, simplify

# Propositions
p, q = symbols('p q')

# Logical Expression
expr = And(p, Or(Not(p), q))

# Simplify the expression
simplified_expr = simplify(expr)
print(f"Simplified Expression: {simplified_expr}")
```

```python
# Truth Table
from sympy.logic.boolalg import truth_table

tt = truth_table(expr, [p, q])
for row in tt:
    print(row)
```

# **Verification of Tautology and Contradiction in Propositional Logic Using Python**
# Objective
To verify if a given compound propositional logic statement is a tautology (true for all cases) or a contradiction (false for all cases) using Python.
# **Theory**
* **Tautology:** A propositional statement is called a tautology if it is true for all possible truth value assignments to its variables.
* **Contradiction:**  A propositional statement is a contradiction if it is false for all possible truth value assignments to its variables.
# **Logical Operators:**
* ¬ (NOT): Negates a proposition.
* ∧ (AND): True if both propositions are true.
* ∨ (OR): True if at least one proposition is true.
* → (Implies): True except when the first proposition is true, and the second is false.
* ↔ (Equivalent): True if both propositions are either true or false.

## **Example 1**

Verify whether the following compound proposition:

 **((P∨Q)∧(¬Q∨R)) →((P∧R)∨(¬Q))**
is a Tautology, Contradiction, or Neither.


---


**Objective**

To verify the logical proposition by constructing a truth table and analyzing its validity under all possible truth value assignments for the propositional variables P, Q, and R.


---
**Theory**

* Logical Operations:

  * OR (∨): True if at least one operand is true.
  * AND (∧): True if both operands are true.
  * NOT (¬): Inverts the truth value of the operand.
  * Implication (→): Defined as ¬A∨B.
* Tautology: A proposition that is always true for all possible truth value assignments.

* Contradiction: A proposition that is always false for all possible truth value assignments.

* Satisfiable: A proposition that is true for at least one truth value assignment.

**Procedure**

* Define the Logical Proposition:

  * The proposition is ((P∨Q)∧(¬Q∨R))→((P∧R)∨(¬Q)).
  * Decompose it into the following parts:
    * LHS:(P∨Q)∧(¬Q∨R).
    * RHS:(P∧R)∨(¬Q).
    * The implication: LHS→RHS, equivalent to ¬LHS∨RHS.
* List Variables:

  * Identify the propositional variables:c

* Create Truth Table:

    * List all possible combinations of truth values for P, Q, R
    * Evaluate each subexpression:
        * P∨Q
        * ¬Q
        * ¬Q∨R
        * LHS:(P∨Q)∧(¬Q∨R)
        * RHS:(P∧R)∨(¬Q)
        * Full Proposition: ((P∨Q)∧(¬Q∨R))→((P∧R)∨(¬Q)).
* Evaluate the Proposition:

    * Analyze the final column in the truth table:
    * If all values are True, the proposition is a Tautology.
    * If all values are False, it is a Contradiction.
    * Otherwise, it is Satisfiable.
* Write Python Code:

    * Use Python to automate the truth table creation and analysis.



# **Observations**

* **Truth Table:** The code generates a table showing all truth value combinations and evaluations for the subexpressions and the full proposition.
* **Final Analysis:**
    * If the final column has all True values, the proposition is a Tautology.
    * If the final column has mixed values, the proposition is Satisfiable.


###**Result**
 For the given proposition:

#  ((P∨Q)∧(¬Q∨R))→((P∧R)∨(¬Q))

the result of the analysis indicates whether it is a Tautology, Contradiction, or Satisfiable.

```python
import pandas as pd
from IPython.display import display

def generate_truth_table(*expressions):
    # Extract variable names from the first expression
    variables = list(expressions[0][1].__code__.co_varnames)
    n = len(variables)  # Number of variables
    rows = 2 ** n  # Number of rows (2^n)

    # Generate all possible combinations of truth values
    data = {var: [(i // (2 ** (n - j - 1))) % 2 == 1 for i in range(rows)] for j, var in enumerate(variables)}

    # Apply logical expressions
    for name, func in expressions:
        data[name] = [func(**{var: row[var] for var in variables}) for row in pd.DataFrame(data).to_dict('records')]

    return pd.DataFrame(data)

# Define the logical proposition
truth_table = generate_truth_table(
    ("(P ∨ Q) ∧ (¬Q ∨ R)", lambda P, Q, R: (P or Q) and (not Q or R)),  # Left-hand side of the implication
    ("(P ∧ R) ∨ (¬Q)", lambda P, Q, R: (P and R) or (not Q)),           # Right-hand side of the implication
    ("((P ∨ Q) ∧ (¬Q ∨ R)) → ((P ∧ R) ∨ (¬Q))", lambda P, Q, R: not ((P or Q) and (not Q or R)) or ((P and R) or (not Q)))  # Full proposition
)

# Display the truth table
display(truth_table)

# Check if the proposition is a tautology
proposition_column = truth_table["((P ∨ Q) ∧ (¬Q ∨ R)) → ((P ∧ R) ∨ (¬Q))"]
if all(proposition_column):
    print("The proposition is a Tautology.")
elif not any(proposition_column):
    print("The proposition is a Contradiction.")
else:
    print("The proposition is Satisfiable.")
```

# **Hasse Diagram of Posets**
A Hasse diagram is a type of diagram used to represent partially ordered sets (Posets). It displays elements as vertices and order relations as edges. The diagram follows specific conventions:

* An edge between two vertices indicates that one element is directly less than another.
* The diagram is drawn such that the elements are arranged in levels, with lower elements at the bottom.

In this example, we will draw a Hasse diagram of the following partially ordered sets (Posets):

* **Poset 1: Divisibility of Numbers**

The *heapq* module in Python provides an implementation of the heap queue algorithm, also known as a priority queue. It is part of Python's standard library and is commonly used for tasks where you need to repeatedly access the smallest (or largest) element in a collection efficiently.

Heaps in *heapq* are implemented as binary heaps and are stored as simple Python lists. They maintain the heap property: the smallest element is always at the root (index 0).

# **Key Features**
* Min-Heap: By default, heapq provides a min-heap, where the smallest element is always at the top.
* Efficiency: Heaps are efficient for operations like insertion and extraction of the smallest element, both of which have O(logn) time complexity.

**Common Functions in heapq**

*heapq.heappush(heap, item)*

Adds an item to the heap while maintaining the heap property.

```python
import heapq
heap = []
heapq.heappush(heap, 5)
heapq.heappush(heap, 3)
heapq.heappush(heap, 8)
print(heap)  # Output: [3, 5, 8]
```

*heapq.heappop(heap)*

Removes and returns the smallest element from the heap.

```python
smallest = heapq.heappop(heap)
print(smallest)  # Output: 3
print(heap)      # Output: [5, 8]
```

*heapq.heappushpop(heap, item)*

Adds a new item to the heap and then removes and returns the smallest item in a single operation.


```python
heap = [3, 5, 8]
result = heapq.heappushpop(heap, 4)
print(result)  # Output: 3
print(heap)    # Output: [4, 5, 8]
```

*heapq.heapreplace(heap, item)*

Removes and returns the smallest element, then adds the new item. This is more efficient than popping and then pushing.


```python
heap = [3, 5, 8]
result = heapq.heapreplace(heap, 4)
print(result)  # Output: 3
print(heap)    # Output: [4, 5, 8]
```

Converts a list into a valid heap in-place.

```python
nums = [5, 3, 8, 1]
heapq.heapify(nums)
print(nums)  # Output: [1, 3, 8, 5]
```

* n is the number of vertices in the graph.
* visited is a boolean list that tracks whether each vertex has been added to the MST.
* mst_edges will store tuples of edges in the MST, e.g., (u, v, weight).
* total_weight accumulates the total weight of the MST.

Initializes a priority queue (min_heap) with a tuple representing the starting vertex:

* weight: 0 (no edge weight to start).
* current_vertex: start.
* previous_vertex: -1 (no previous vertex since it's the start).

* The algorithm processes edges in order of increasing weight by popping the smallest edge from min_heap.
* If the vertex u has already been visited, it skips to the next edge.
* Marks the vertex as visited and adds the edge's weight to total_weight

If prev (the previous vertex) is valid, the edge (prev, u) is added to the MST.

    * Iterates over all vertices v adjacent to u.
    * Adds the edge (w, v, u) to min_heap if vertex v is unvisited and the edge weight w is positive.

Returns the list of MST edges and the total weight of the MST.

```python
import heapq
import matplotlib.pyplot as plt
import networkx as nx

# ------------------- Prim's Algorithm -------------------
def prim_mst(graph, start=0):
    n = len(graph)  # Step 1: Count the number of vertices
    visited = [False] * n  # Step 2: Track visited nodes
    mst_edges = []         # Step 3: Store MST edges
    total_weight = 0       # Step 4: Store total weight of MST

    # Step 5: Min-heap to select smallest edge at each step (weight, to_node, from_node)
    min_heap = [(0, start, -1)]  # Step 6: Start from the source node (start)

    # Step 7: Process heap until all nodes are visited
    while min_heap:
        weight, u, prev = heapq.heappop(min_heap)  # Step 8: Get edge with min weight
        if visited[u]:  # Step 9: Skip if already visited
            continue
        visited[u] = True  # Step 10: Mark node as visited
        total_weight += weight  # Step 11: Add edge weight to total

        if prev != -1:  # Step 12: If not the starting node, add edge to MST
            mst_edges.append((prev, u, weight))

        # Step 13: Push all valid (non-visited) neighboring edges into heap
        for v, w in enumerate(graph[u]):
            if not visited[v] and w > 0:
                heapq.heappush(min_heap, (w, v, u))

    return mst_edges, total_weight

# ------------------- Kruskal's Algorithm -------------------
class DisjointSet:
    def __init__(self, n):
        self.parent = list(range(n))  # Step 1: Initialize parent to self
        self.rank = [0] * n           # Step 2: Initialize rank for union by rank

    def find(self, u):
        if self.parent[u] != u:      # Step 3: Path compression
            self.parent[u] = self.find(self.parent[u])
        return self.parent[u]

    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:  # Step 4: Union by rank to keep tree shallow
            if self.rank[root_u] > self.rank[root_v]:
                self.parent[root_v] = root_u
            elif self.rank[root_u] < self.rank[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.rank[root_u] += 1

def kruskal_mst(edges, n):
    ds = DisjointSet(n)  # Step 1: Create disjoint set for 'n' vertices
    mst_edges = []       # Step 2: List to store MST edges
    total_weight = 0     # Step 3: Track total MST weight

    # Step 4: Sort all edges by weight
    edges.sort(key=lambda x: x[2])

    # Step 5: Iterate through sorted edges
    for u, v, weight in edges:
        # Step 6: Add edge if it doesn't form a cycle
        if ds.find(u) != ds.find(v):
            ds.union(u, v)
            mst_edges.append((u, v, weight))
            total_weight += weight

    return mst_edges, total_weight

# ------------------- Visualization Function -------------------
def plot_graph_and_mst(graph, mst_edges, title):
    G = nx.Graph()

    # Step 1: Build graph from adjacency matrix
    for i, row in enumerate(graph):
        for j, weight in enumerate(row):
            if weight > 0:
                G.add_edge(i, j, weight=weight)

    pos = nx.spring_layout(G, seed=42)  # Step 2: Layout positions

    # Step 3: Draw base graph
    plt.figure(figsize=(10, 8))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)})

    # Step 4: Highlight MST edges in red
    mst_edges_set = set((u, v) for u, v, _ in mst_edges)
    mst_edges_set |= set((v, u) for u, v, _ in mst_edges)
    nx.draw_networkx_edges(G, pos, edgelist=mst_edges_set, edge_color='red', width=2)

    plt.title(title)
    plt.show()

# ------------------- Example Input and Execution -------------------

# Step 1: Adjacency matrix for Prim's algorithm
graph = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 8, 5],
    [0, 3, 0, 0, 7],
    [6, 8, 0, 0, 9],
    [0, 5, 7, 9, 0]
]

# Step 2: Run Prim's Algorithm
prim_mst_edges, prim_total_weight = prim_mst(graph)
print("Prim's MST Edges:", prim_mst_edges)
print("Total Weight:", prim_total_weight)
plot_graph_and_mst(graph, prim_mst_edges, "Prim's Algorithm MST")

# Step 3: Edge list for Kruskal's algorithm
edges = [
    (0, 1, 2), (1, 2, 3), (0, 3, 6),
    (1, 3, 8), (1, 4, 5), (2, 4, 7),
    (3, 4, 9)
]
n = 5

# Step 4: Convert edges to adjacency matrix for plotting
adj_matrix = [[0] * n for _ in range(n)]
for u, v, w in edges:
    adj_matrix[u][v] = w
    adj_matrix[v][u] = w

# Step 5: Run Kruskal's Algorithm
kruskal_mst_edges, kruskal_total_weight = kruskal_mst(edges, n)
print("Kruskal's MST Edges:", kruskal_mst_edges)
print("Total Weight:", kruskal_total_weight)
plot_graph_and_mst(adj_matrix, kruskal_mst_edges, "Kruskal's Algorithm MST")
```

# **Dijkstra’s Algorithm for Shortest Path**
**Objective**

Understand and implement Dijkstra's Algorithm to find the shortest path from a source node to all other nodes in a weighted graph.

**Problem Statement**

Write a Python program that uses Dijkstra’s algorithm to compute the shortest path from a given source node to all other nodes in a weighted, connected, and directed graph. Output the shortest path distances.

# **Algorithm Steps**
* Initialize the distance of the source node as 0 and all other nodes as infinity.
* Create a priority queue (min-heap) to store nodes and their current shortest distances.
* While the priority queue is not empty:
    * Extract the node with the minimum distance.
    * Update the distance for its neighboring nodes.
    * Push the neighbors with updated distances into the queue.
* Repeat until all nodes have been visited or the queue is empty.

# **Dijkstra’s Algorithm with Visualization**
**Objective**

* Implement Dijkstra's Algorithm in Python to find the shortest path.
* Visualize the graph and shortest path distances.

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt

# ------------------- Dijkstra's Algorithm -------------------
def dijkstra(graph, start):
    n = len(graph)  # Step 1: Total number of nodes
    distances = [float('inf')] * n  # Step 2: Initialize distances to ∞
    distances[start] = 0            # Step 3: Distance to source is 0
    priority_queue = [(0, start)]   # Step 4: Min-heap for selecting shortest tentative distance
    previous_nodes = [None] * n     # Step 5: Track path back to source

    # Step 6: Continue while there are unvisited nodes in the priority queue
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Step 7: Skip processing if we have already found a shorter path
        if current_distance > distances[current_node]:
            continue

        # Step 8: Check neighbors of the current node
        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight  # Tentative distance via current node

            # Step 9: Update if a shorter path is found
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_nodes[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))  # Add to heap

    return distances, previous_nodes

# ------------------- Visualization -------------------
def plot_graph(graph, shortest_distances, start_node):
    G = nx.DiGraph()

    # Step 1: Add nodes and edges to NetworkX graph
    for node, neighbors in graph.items():
        for neighbor, weight in neighbors:
            G.add_edge(node, neighbor, weight=weight)

    pos = nx.spring_layout(G)  # Step 2: Layout for nodes

    # Step 3: Draw nodes and edges
    edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
    nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=14, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Step 4: Highlight shortest paths in red
    for i, distance in enumerate(shortest_distances):
        if i != start_node and distance < float('inf'):
            path = reconstruct_path(i, start_node, previous_nodes)
            highlight_edges = [(path[j], path[j + 1]) for j in range(len(path) - 1)]
            nx.draw_networkx_edges(G, pos, edgelist=highlight_edges, edge_color='red', width=2.5)

    plt.title(f"Graph with Shortest Paths from Node {start_node}")
    plt.show()

# ------------------- Path Reconstruction -------------------
def reconstruct_path(target_node, start_node, previous_nodes):
    path = []
    current = target_node
    # Step 1: Backtrack from target to start using previous_nodes
    while current is not None:
        path.append(current)
        current = previous_nodes[current]
    path.reverse()  # Step 2: Reverse to get path from start to target
    return path

# ------------------- Example Graph -------------------
# Step 1: Graph is defined using adjacency list
graph = {
    0: [(1, 4), (2, 1)],
    1: [(3, 1)],
    2: [(1, 2), (3, 5)],
    3: []
}

# Step 2: Run Dijkstra from source node 0
source_node = 0
shortest_distances, previous_nodes = dijkstra(graph, source_node)

# Step 3: Display shortest distances
print("Shortest distances from node", source_node, "to all other nodes:", shortest_distances)

# Step 4: Visualize the graph and shortest paths
plot_graph(graph, shortest_distances, source_node)
```

# **Graph Coloring using Greedy Algorithm**
**Objective**

* Understand and implement the Greedy Algorithm for graph coloring.
* Visualize the graph with colored nodes.

 **Problem Statement**

Write a Python program to color the nodes of a given graph using the Greedy Graph Coloring algorithm. The program should minimize the number of colors while ensuring that no two adjacent nodes share the same color.

# **Algorithm Steps**
* Sort the nodes in any order (in the Greedy approach, the order can affect the result).
* Assign the smallest possible color to each node such that no two adjacent nodes have the same color.
* Track the color assignments to ensure the validity of the solution.

```python
import networkx as nx
import matplotlib.pyplot as plt

# ------------------- Graph Coloring Function -------------------
def greedy_graph_coloring(graph):
    # Step 1: Initialize a dictionary to store colors for each node
    color_assignment = {}

    # Step 2: Loop through each node in the graph
    for node in graph.nodes():
        # Step 3: Get the colors of all adjacent (neighboring) nodes
        adjacent_colors = {color_assignment.get(neighbor) for neighbor in graph.neighbors(node)}

        # Step 4: Assign the smallest possible color (integer) that isn't taken by a neighbor
        color = 0
        while color in adjacent_colors:
            color += 1
        color_assignment[node] = color

    return color_assignment

# ------------------- Visualization Function -------------------
def plot_colored_graph(graph, color_assignment):
    # Step 1: Choose positions for the nodes
    pos = nx.spring_layout(graph)

    # Step 2: Retrieve assigned colors from the dictionary
    colors = [color_assignment[node] for node in graph.nodes()]

    # Step 3: Draw the graph with node colors based on the color assignment
    nx.draw(
        graph, pos, with_labels=True,
        node_color=colors, cmap=plt.cm.rainbow,
        node_size=700, font_size=14, font_weight='bold',
        edge_color='black'
    )

    plt.title("Graph Coloring using Greedy Algorithm")
    plt.show()

# ------------------- Create Sample Graph -------------------
# Step 1: Define edges of the graph (conflicting pairs)
edges = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4)]

# Step 2: Create a NetworkX graph and add edges
G = nx.Graph()
G.add_edges_from(edges)

# ------------------- Run Greedy Coloring -------------------
color_assignment = greedy_graph_coloring(G)

# Step 3: Print color assignment result
print("Node Color Assignment:", color_assignment)

# Step 4: Plot the colored graph
plot_colored_graph(G, color_assignment)
```
