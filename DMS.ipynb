# **2. SymPy**

**Purpose:** Symbolic mathematics, logic manipulation, and discrete algebra.

**Applications:**

	* Logical equivalences
	* Truth tables
	* Propositional calculus


```python
from sympy import symbols, And, Or, Not, Implies, simplify

# Propositions
p, q = symbols('p q')

# Logical Expression
expr = And(p, Or(Not(p), q))

# Simplify the expression
simplified_expr = simplify(expr)
print(f"Simplified Expression: {simplified_expr}")
```

```python
# Truth Table
from sympy.logic.boolalg import truth_table

tt = truth_table(expr, [p, q])
for row in tt:
    print(row)
```

```python
from sympy import symbols, simplify_logic

# Logical expression simplification
A, B = symbols('A B')
expr = (A & ~B) | (~A & B)  # XOR operation
simplified_expr = simplify_logic(expr)
print("Simplified Expression:", simplified_expr)

# Truth Table
from sympy.logic.boolalg import truth_table

tt = truth_table(expr, [A, B])
for row in tt:
    print(row)
```

# **Aim:**
To develop a Python program that generates and displays truth tables for multiple logical expressions involving Boolean variables.

# **Objectives:**

	*  Understand the representation of logical expressions as Python functions.
	*  Implement a function to generate all possible combinations of truth values for given variables.
	*  Evaluate multiple logical expressions for each combination of inputs.
	*  Display the resulting truth table in a readable tabular format.
	*  Render mathematical logical formulas using LaTeX for clarity.


# **Algorithm:**

**Import necessary libraries:**

*	Use *pandas* for table creation
*	***Python.display*** for LaTeX rendering in notebooks.

**Explanation:**
* pandas: This library is used to handle and manipulate tabular data, and here it's used to create and display the truth table in a DataFrame format.
* IPython.display: This module provides a way to render LaTeX in Jupyter environments. Math is used for rendering mathematical formulas.


```python
import pandas as pd
from IPython.display import display, Math
```


 **Define function:**
*	**generate\_truth\_table(expressions)** to handle multiple logical expressions dynamically.
* **Extract Variables:** From the first logical expression's function, extract the list of variable names using Python's *func.\_\_code\_\_.co\_varnames*. Assumes all expressions use the same set of variables.

**Calculate Number of Rows**

			* Let **n** be the number of Boolean variables.
			* Compute the total number of rows as **2^n**, which represents all possible combinations of input values.

**Generate Truth Value Combinations**

			* For each variable **var ** at position **j** (0-indexed):
			* Generate a list of Boolean values of length **2^n** such that the truth value toggles every **2^{n - j - 1}** rows.
			* For example, the first variable toggles every half of the rows, the second every quarter, and so on.
			* Store these lists in a dictionary with variable names as keys.

**Create Initial Data Table**
			* Combine all generated truth value lists into a data structure (e.g., a dictionary).
			* This structure will represent the variable columns of the truth table.
**Evaluate Logical Expressions:** For each logical expression **name**, **func**in the input:
				* For each row (combination of variable truth values), call **func** with the corresponding arguments.
				* Store the result in the data structure under the column labeled with **name**.

**Return or Display Result**

			* Convert the final data structure into a formatted table using a Python tool like **pandas.DataFrame**.
			* Return or display the truth table in readable tabular format.

```python
def generate_truth_table(*expressions):
    # Extract variable names from the first lambda function's argument list
    variables = list(expressions[0][1].__code__.co_varnames)
    # Count how many variables there are (e.g., 2 for A, B)
    n = len(variables)
    # Compute number of truth table rows (2^n combinations)
    rows = 2 ** n

    # Generate all possible truth value combinations for each variable
    # For each variable (indexed by j), we generate True/False for all rows
    # The formula gives us a binary pattern for that column
    data = {
        var: [(i // (2 ** (n - j - 1))) % 2 == 1 for i in range(rows)]
        for j, var in enumerate(variables)
    }

    # Now evaluate each logical expression over all generated truth rows
    for name, func in expressions:
        # For each expression, apply the function row-by-row
        # `row[var] for var in variables` gives the inputs to the function
        data[name] = [
            func(**{var: row[var] for var in variables})
            for row in pd.DataFrame(data).to_dict('records')
        ]

    # Convert the full dataset (variables + logical expressions) to a DataFrame
    return pd.DataFrame(data)
```

# **Verification of Tautology and Contradiction in Propositional Logic Using Python**

---


---



---




# **Aim:**

To develop a Python program that generates and displays

 truth tables for multiple logical expressions involving

 Boolean variables.

# **Objectives:**

* Understand the representation of logical expressions as Python functions.
* Implement a function to generate all possible combinations of truth values

 for given variables.
* Evaluate multiple logical expressions for each combination of inputs.
* Display the resulting truth table in a readable tabular format.
* Render mathematical logical formulas using LaTeX for clarity.





# **Algorithm**

  * Import required libraries: pandas} and IPython.display}.
  * Declare propositional variables: let,$P$, $Q$, $R$.
  * Construct the logical expression.
#    **EXAMPLE**
    * LHS: $(P \lor Q) \land (\lnot Q \lor R)$
    * RHS: $(P \land R) \lor \lnot Q$
    * Full: LHS $\rightarrow$ RHS
  * Generate all $2^3 = 8$ combinations of truth values.
  * Evaluate each row.
  * Check if the full expression is always True (tautology), always False (contradiction), or mixed (satisfiable).

```python
import pandas as pd
from IPython.display import display

def generate_truth_table(*expressions):
    # Extract variable names from the first expression
    variables = list(expressions[0][1].__code__.co_varnames)
    n = len(variables)  # Number of variables
    rows = 2 ** n  # Number of rows (2^n)

    # Generate all possible combinations of truth values
    data = {var: [(i // (2 ** (n - j - 1))) % 2 == 1
                  for i in range(rows)] for j, var in enumerate(variables)}

    # Apply logical expressions
    for name, func in expressions:
        data[name] = [func(**{var: row[var] for var in variables})
         for row in pd.DataFrame(data).to_dict('records')]

    return pd.DataFrame(data)

# Define the logical proposition
truth_table = generate_truth_table(
    ("(P ∨ Q) ∧ (¬Q ∨ R)", lambda P, Q, R: (P or Q) and (not Q or R)),  # Left-hand side of the implication
    ("(P ∧ R) ∨ (¬Q)", lambda P, Q, R: (P and R) or (not Q)),           # Right-hand side of the implication
    ("((P ∨ Q) ∧ (¬Q ∨ R)) → ((P ∧ R) ∨ (¬Q))", lambda P, Q, R:
     not ((P or Q) and (not Q or R)) or ((P and R) or (not Q)))  # Full proposition
)

# Display the truth table
display(truth_table)
```

```python
# Check if the proposition is a tautology
proposition_column = truth_table["((P ∨ Q) ∧ (¬Q ∨ R)) → ((P ∧ R) ∨ (¬Q))"]
if all(proposition_column):
    print("The proposition is a Tautology.")
elif not any(proposition_column):
    print("The proposition is a Contradiction.")
else:
    print("The proposition is Not Tautology Nor Contradiction .")
```

```python
# Define the logical proposition
truth_table = generate_truth_table(
    ("LHS: (P ∨ Q) ∧ (¬P ∨ R)", lambda P, Q, R: (P or Q) and (not P or R)),  # Left-hand side of the implication
    ("RHS: Q ∨ R", lambda P, Q, R: Q or R),                                  # Right-hand side of the implication
    ("Proposition: ((P ∨ Q) ∧ (¬P ∨ R)) → (Q ∨ R)", lambda P, Q, R: not ((P or Q) and (not P or R)) or (Q or R))  # Full proposition
)

# Display the truth table
display(truth_table)

# Check if the proposition is a tautology
proposition_column = truth_table["Proposition: ((P ∨ Q) ∧ (¬P ∨ R)) → (Q ∨ R)"]
if all(proposition_column):
    print("The proposition is a Tautology.")
elif not any(proposition_column):
    print("The proposition is a Contradiction.")
else:
    print("The proposition is Satisfiable.")
```

```python
# Define the logical proposition
truth_table = generate_truth_table(
    ("P", lambda P: P),                     # The truth value of P
    ("¬P", lambda P: not P),                # The negation of P
    ("Proposition: P ∧ ¬P", lambda P: P and not P)  # The full contradiction
)

# Display the truth table
display(truth_table)

# Check if the proposition is a contradiction
proposition_column = truth_table["Proposition: P ∧ ¬P"]
if all(proposition_column):
    print("The proposition is a Tautology.")
elif not any(proposition_column):
    print("The proposition is a Contradiction.")
else:
    print("The proposition is Satisfiable.")
```

# **Aim**
To visualize partially ordered sets (Posets) using Hasse diagrams for better understanding of order relations like divisibility and subset relations.

# **Objectives**

*  To construct and visualize the Hasse diagram for:
    * A set of numbers with the divisibility relation.
    * A set of subsets with the subset relation.
* To understand hierarchical representation in partially ordered sets.
* To implement Hasse diagram construction using Python libraries: *networkx* and *matplotlib*.



Algorithm: Hasse Diagram Construction

1. Input:
   - A finite set S
   - A partial order relation <= on S

2. Build covering relations:
   For each pair (a, b) in S with a != b:
       If a <= b and there is no c in S with a < c < b:
           Add edge (a, b) to the Hasse diagram.

3. Determine levels:
   - Minimal elements (those with no predecessors) are assigned to level 0.
   - For each other element x:
       level(x) = max{ level(y) + 1 | (y, x) is an edge }.

4. Arrange nodes:
   - Place nodes in rows according to their level.
   - Spread nodes horizontally within the same level.
**Layout for Visualization:**

      * Place elements of the same level in a horizontal row.

      * P lace higher-level elements above lower-level ones.

      * Draw edges only between levels (no diagonals crossing intermediate nodes).
5. Draw edges:
   - For each covering relation (a, b), draw a line from a (lower level) to b (higher level).
   - Edges are drawn upward, arrowheads optional.

Output:
   - A layered diagram with minimal elements at the bottom,
     maximal elements at the top, and only covering relations shown.


# **Example 1: Divisibility of Numbers Poset**
Consider the set S={1,2,3,4,6,12} with the divisibility relation. This means that element a is less than element b if a divides b.

```python
import itertools
import networkx as nx

def hasse_edges(S, relation):
    """
    Compute Hasse edges for a poset.
    - S: list of elements
    - relation(a, b): returns True if a < b in the poset
    """
    edges = []
    for a, b in itertools.permutations(S, 2):
        if relation(a, b):  # a < b
            # Check if there's no c with a < c < b
            if not any(relation(a, c) and relation(c, b) for c in S if c not in (a, b)):
                edges.append((a, b))
    return edges

# Example: divisibility poset
S = [1, 2, 3, 4, 6, 12]
relation = lambda a, b: (b % a == 0 and a != b)

H_edges = hasse_edges(S, relation)
print("Hasse edges:", H_edges)

# Build Hasse diagram
H = nx.DiGraph()
H.add_nodes_from(S)
H.add_edges_from(H_edges)

import matplotlib.pyplot as plt
pos = nx.spring_layout(H)      # force-directed (default, "organic")
# pos = nx.circular_layout(H)    # nodes arranged in a circle
# pos = nx.planar_layout(H)
plt.figure(figsize=(8, 6))
nx.draw(H, pos, with_labels=True, node_size=3000, node_color="lightblue",
        font_size=12, font_weight="bold", arrows=True)
plt.title("Hasse Diagram (automatically computed)", fontsize=14)
plt.show()
```

# **Example 2: Subsets of a Set Poset**
Consider the set S={a,b,c}, with the subset relation. This means element A is less than element B if A is a subset of B.

# **Algorithm for Constructing a Hasse Diagram**

 Step 1: Start with a finite set S.
 Example: S = {a, b, c}

 Step 2: Define the partial order relation on S.
 Example: for subsets, relation(a, b) means "a is a proper subset of b".

 Step 3: Generate all possible pairs (a, b) from S.
 Check if a < b according to the relation.

 Step 4: For each such pair (a, b),
 verify if there exists an element c such that a < c < b.
 If such c exists, then (a, b) is not a covering relation.

 Step 5: Keep only covering relations (those pairs with no element in between).
 These become the edges of the Hasse diagram.

 Step 6: Place nodes (elements of S) into levels according to their "rank"
 (for subsets: by their size, for divisibility: by number of prime factors, etc.).

 Step 7: Draw or list the diagram:
   - Nodes: all elements of S
   - Edges: covering relations only


```python
from itertools import combinations, permutations

# ---------- Helper Functions ----------
def powerset(iterable):
    """Return power set as frozensets (hashable for graph)."""
    s = list(iterable)
    return [frozenset(c) for r in range(len(s)+1) for c in combinations(s, r)]

def hasse_edges(S, relation):
    """Compute covering relations (Hasse edges)."""
    edges = []
    for a, b in permutations(S, 2):
        if relation(a, b):
            # keep edge only if no c exists with a < c < b
            if not any(relation(a, c) and relation(c, b) for c in S if c not in (a, b)):
                edges.append((a, b))
    return edges

# Example 2: Subsets of {a, b, c}
U = {"a", "b", "c"}
S = powerset(U)

# Relation = proper subset
relation = lambda a, b: a < b

edges = hasse_edges(S, relation)
# Print powerset nicely
print("Power set of {a,b,c}:")
for subset in S:
    print(set(subset))

# Print Hasse edges nicely
print("\nHasse edges:")
for a, b in edges:
    print(set(a), " < ", set(b))

# Build graph
G = nx.DiGraph()
G.add_nodes_from(S)
G.add_edges_from(edges)

# Assign levels based on subset size
levels = {node: len(node) for node in S}
pos = {node: (i, levels[node]) for i, node in enumerate(S)}

# Draw graph
plt.figure(figsize=(20, 15))
nx.draw(G, pos, with_labels=True, node_size=1500, node_color="lightblue",
        font_size=10, font_weight="bold", arrows=False,
        labels={n: "{" + ",".join(n) + "}" if n else "∅" for n in G.nodes()})
plt.title("Hasse Diagram of Subsets of {a,b,c}")
plt.show()
```

# **Example simple tree**

**Explanation of the Code:**

* Tree Structure (Parent-Child Relations):

    * The tree structure is defined by a list of parent-child pairs: [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7)].

* Manual Node Positioning:

    * We manually set the positions of the nodes in a dictionary (pos) to ensure the tree structure is maintained:
      * Node 1 is at the top of the tree.
      * Nodes 2 and 3 are directly beneath node 1, and so on.
      * Nodes 4, 5, 6, and 7 are placed as leaves of the tree.

* Drawing the Tree:

    * The tree is drawn using networkx's draw function, with arrows indicating the direction of the edges (from parent to child).

* Visualization:

    * The layout positions the nodes in a tree-like manner, which reflects the exact hierarchical structure.

* **Expected Output:**

  * The resulting plot will display the exact tree structure where:
    * Node 1 is at the top (root).
    * Nodes 2 and 3 are directly below node 1.
    * Nodes 4, 5, 6, and 7 are the leaf nodes placed at the lowest level.

This approach guarantees that the tree is visualized accurately, with a clear hierarchical structure.








```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to draw the exact tree structure (Hasse diagram)
def draw_exact_tree(tree_structure):
    G = nx.DiGraph()  # Directed graph for Poset (Hasse diagram)

    # Add edges based on parent-child relations
    G.add_edges_from(tree_structure)

    # Manually set node positions to reflect the tree structure
    pos = {
        1: (0, 3),
        2: (-1, 2),
        3: (1, 2),
        4: (-1.5, 1),
        5: (-0.5, 1),
        6: (0.5, 1),
        7: (1.5, 1)
    }

    plt.figure(figsize=(8, 6))

    # Draw the tree graph
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightgreen', font_size=12, font_weight="bold", arrows=True)

    # Display the title
    plt.title("Exact Tree Structure as Hasse Diagram", fontsize=16)
    plt.show()

# Simple tree structure: (parent, child) pairs
tree_structure = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7)]

# Draw the exact tree structure
draw_exact_tree(tree_structure)
```

#**Algorithm for Plotting a Directed Graph**

	* Import required libraries: networkx and matplotlib.pyplot.
	* Create a directed graph object using DG = nx.DiGraph().
	* Add nodes and directed edges using DG.add\_edges\_from() with ordered pairs.

		* with\_labels=True for node labels.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Creating a directed graph
DG = nx.DiGraph()

# Add nodes and edges
DG.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

# Plotting the directed graph with proper labeling
plt.figure(figsize=(6, 6))
nx.draw(DG, with_labels=True, node_size=3000, node_color='lightblue', font_size=14, font_weight='bold', arrows=True )

# Title for the plot
plt.title('Directed Graph Example', fontsize=16)
plt.show()
```
