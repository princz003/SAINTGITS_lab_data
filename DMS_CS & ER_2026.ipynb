# 1. **SymPy**

**Purpose:**

symbolic logic manipulation, simplifying Boolean expressions, generating truth tables. Below we simplify p ∧ (¬p ∨ q) and confirm it reduces to p ∧ q.


```python
from sympy import symbols, And, Or, Not, Implies, simplify

# Propositions
p, q = symbols('p q')

# Logical Expression
expr = And(p, Or(Not(p), q))

# Simplify the expression
simplified_expr = simplify(expr)
print(f"Simplified Expression: {expr}")
print(f"Simplified Expression: {simplified_expr}")
```

```python
# Truth Table
from sympy.logic.boolalg import truth_table

tt = truth_table(expr, [p, q])
for row in tt:
    #print(row)
    display(row)
```

# **Note:**

Now build the truth table for the XOR expression (A ∧ ¬B) ∨ (¬A ∧ B) and confirm SymPy leaves it unchanged (it's already minimal).

```python

from sympy import symbols, simplify_logic

# Logical expression simplification
A, B = symbols('A B')
expr = (A & ~B) | (~A & B)  # XOR operation
simplified_expr = simplify_logic(expr)
print(f"Simplified Expression: {expr}")
print("Simplified Expression:", simplified_expr)
```

```python
# Truth Table
from sympy.logic.boolalg import truth_table

tt = truth_table(expr, [A, B])
for row in tt:
    print(row)
```

# **3. NetworkX**
**Purpose:**

build graphs, test connectivity, and find shortest paths. Below is a 4-cycle graph (1-2-3-4-1); we check it's connected and find the shortest path from 1 to 3.




```python
import networkx as nx

# Create a graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Check connectivity
print(f"Is graph connected? {nx.is_connected(G)}")

# Find shortest path
path = nx.shortest_path(G, source=1, target=3)
print(f"Shortest path from 1 to 3: {path}")
```

# **Constructing a truth table of different compound statements in propositional logic.**

# **Step 1: Import Required Libraries**

* *pandas:* This library is used to handle and manipulate tabular data, and here it's used to create and display the truth table in a DataFrame format.
* *IPython.display*: This module provides a way to render LaTeX in Jupyter environments. Math is used for rendering mathematical formulas.



```python
import pandas as pd
from IPython.display import display, Math
```

# **Step 2: Define the generate_truth_table Function**

**Step-by-Step Explanation of the Code in function**

---


**Step 2.1: Define the function**

* generate_truth_table(*expressions): This function takes multiple compound logical expressions as input. The *expressions parameter allows you to pass any number of arguments, where each argument is a tuple of the form (name, function).

**Step 2.2: Extract variables and define the number of rows**
* variables = list(expressions[0][1].__code__.co_varnames): Extracts the variable names used in the first logical function (using co_varnames from the function's code).
* n = len(variables): Calculates the number of variables involved in the logical expressions.
* rows = 2 ** n: Calculates the number of rows in the truth table, which is
2^n, where, 'n' is the number of variables.

**Step 2.3: Generate truth values for each variable**

The data dictionary is populated with truth values for each variable. The dictionary comprehension loops through the variables and generates all possible combinations of truth values (True or False) for them.

---


**Step 2.5: Return the truth table**

After calculating all logical expressions, the data dictionary is converted into a Pandas DataFrame and returned.


```python
def generate_truth_table(*expressions):
    # Get variable names from the first function
    variables = list(expressions[0][1].__code__.co_varnames)
    n = len(variables)
    rows = 2 ** n  # Number of rows in the truth table

    # Generate truth values for each variable
    data = {var: [(i // (2 ** (n - j - 1))) % 2 == 1 for i in range(rows)] for j, var in enumerate(variables)}

    # Compute the values of each expression
    for name, func in expressions:
        data[name] = [func(**{var: row[var] for var in variables}) for row in pd.DataFrame(data).to_dict('records')]

    # Return the truth table as a Pandas DataFrame
    return pd.DataFrame(data)
```


# **Step 3: Define Compound Statements**


```python
truth_table = generate_truth_table(
    ("A ∨ B", lambda A, B: A or B),             # Logical OR
    ("A ∧ B", lambda A, B: A and B),            # Logical AND
    ("¬A", lambda A, B: not A),                 # Logical NOT
    ("A → B", lambda A, B: not A or B),         # Implication
    ("A ↔ B", lambda A, B: A == B),             # Biconditional
)
```

# ***Example 2***

Using the generate_truth_table() function shown above, construct the truth table for a login system's access rule A ∧ (B ∨ C), where A = "username is valid", B = "password is correct", and C = "valid OTP provided", and identify all input combinations under which access is granted.

```python
import pandas as pd
from IPython.display import display, Math

def generate_truth_table(*expressions):
    variables = list(expressions[0][1].__code__.co_varnames)  # Extract variable names
    n = len(variables)  # Number of variables
    rows = 2 ** n  # Number of rows (2^n)

    # Generate all possible combinations of truth values for variables
    data = {var: [(i // (2 ** (n - j - 1))) % 2 == 1 for i in range(rows)] for j, var in enumerate(variables)}

    # Apply each logical expression function
    for name, func in expressions:
        data[name] = [func(**{var: row[var] for var in variables}) for row in pd.DataFrame(data).to_dict('records')]

    return pd.DataFrame(data)

# Define compound logical expressions
truth_table = generate_truth_table(
    ("A ∧ (B ∨ C)", lambda A, B, C: A and (B or C)),  # Logical AND and OR
    ("(A → B) ∧ (B → C)", lambda A, B, C: (not A or B) and (not B or C)),  # Logical AND of implications
    ("¬A ∨ ¬B", lambda A, B, C: not A or not B)  # Logical OR of negations
)

# Display the truth table
display(truth_table)
```

# **Result**
Python programs using pandas and IPython.display generated truth tables for compound propositional statements automatically: five 2-variable expressions (A∨B, A∧B, ¬A, A→B, A↔B; 4 rows) and three 3-variable expressions (A∧(B∨C), (A→B)∧(B→C), ¬A∨¬B; 8 rows). All outputs matched expected logical identities on manual verification (e.g., A→B is False only when A=True, B=False; A↔B is True exactly when A and B agree).

# **Discussion**
The function generalises to any number of variables by reading __code__.co_varnames from the first lambda, so a new expression needs no code changes — only a new (name, function) pair. This scales as O(2ⁿ) rows, which is exactly why automating truth tables matters once n ≥ 3: it replaces error-prone manual enumeration.
Separating "logic" (the lambda) from "truth-table machinery" (generate_truth_table) lets one function handle OR, AND, NOT, implication, biconditional, and arbitrary nested combinations unchanged — the same exhaustive-evaluation principle used by SAT solvers and hardware verification tools to prove or disprove properties like "this gate is always True" or "this access rule never over-grants."
